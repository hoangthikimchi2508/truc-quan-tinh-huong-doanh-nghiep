#### <font color='blue violet'>2.2.2.8 </font>

#### <font color='blue violet'>2.2.2.9 </font>

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
labels = ['1948', '1949', '1950', '1951', '1952']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    plt.plot(unrate['MONTH'][start_index:end_index], unrate['VALUE'][start_index:end_index],
             c=colors[i], label=labels[i])

    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)
# Đặt tiêu đề và nhãn cho các trục
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.legend(loc='upper left')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulating a DataFrame similar to the 'recent_grads' DataFrame
# with 'Major', 'ShareWomen', and 'Unemployment_rate' columns and 20 rows of sample data
data = {
    'Major': [f'Major {i}' for i in range(20)],
    'ShareWomen': [round(0.1 * i, 2) for i in range(20)],
    'Unemployment_rate': [round(0.05 * i, 2) for i in range(20)]
}
recent_grads = pd.DataFrame(data)

# Plotting bar plots as per the instructions

# 1. Use bar plots to compare the percentages of women (ShareWomen) from the first ten rows and last ten rows
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

recent_grads.head(10).plot.bar(x='Major', y='ShareWomen', ax=ax[0], legend=False)
ax[0].set_title('Percentage of Women in the First Ten Majors')

recent_grads.tail(10).plot.bar(x='Major', y='ShareWomen', ax=ax[1], legend=False)
ax[1].set_title('Percentage of Women in the Last Ten Majors')

# Adjust layout
plt.tight_layout()
plt.show()

# 2. Use bar plots to compare the unemployment rate (Unemployment_rate) from the first ten rows and last ten rows
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

recent_grads.head(10).plot.bar(x='Major', y='Unemployment_rate', ax=ax[0], legend=False)
ax[0].set_title('Unemployment Rate in the First Ten Majors')

recent_grads.tail(10).plot.bar(x='Major', y='Unemployment_rate', ax=ax[1], legend=False)
ax[1].set_title('Unemployment Rate in the Last Ten Majors')

# Adjust layout
plt.tight_layout()
plt.show()
```
