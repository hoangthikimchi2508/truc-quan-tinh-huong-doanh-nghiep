```python
# Major categories to plot
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']

# Creating a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

# Loop through the major categories and create a subplot for each
for sp in range(4):
    ax = axes[sp//2, sp%2]  # Locate the right subplot
    # Plot women's percentages
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], color='blue', label='Women')
    # Plot men's percentages (assuming it's the complement to 100% of the women's percentage)
    ax.plot(women_degrees['Year'], 100 - women_degrees[major_cats[sp]], color='green', label='Men')
    # Set x and y axis limits
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    # Hide all of the spines and tick marks
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False, labelleft=False)
    for spine in ax.spines.values():
        spine.set_visible(False)
    # Set title of each subplot to the name of the major category
    ax.set_title(major_cats[sp])

# Add a legend to the last subplot (bottom right subplot)
axes[-1, -1].legend(loc='upper right')

# Display the plot
plt.show()
```

```python
# Major categories to plot
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']

# RGB values for the colorblind-friendly palette
women_color = (0, 107, 164)  # Dark blue
men_color = (255, 128, 14)   # Orange

# Normalize the RGB values to [0, 1] since Matplotlib expects these values
women_color = tuple([x/255.0 for x in women_color])
men_color = tuple([x/255.0 for x in men_color])

# Creating a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

# Loop through the major categories and create a subplot for each
for sp in range(4):
    ax = axes[sp//2, sp%2]  # Locate the right subplot
    # Plot women's percentages with the new color
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], color=women_color, label='Women' if sp==0 else "_nolegend_")
    # Plot men's percentages with the new color
    ax.plot(women_degrees['Year'], 100 - women_degrees[major_cats[sp]], color=men_color, label='Men' if sp==0 else "_nolegend_")
    # Set x and y axis limits
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    # Hide all of the spines and tick marks
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False, labelleft=False)
    for spine in ax.spines.values():
        spine.set_visible(False)
    # Set title of each subplot to the name of the major category
    ax.set_title(major_cats[sp])

# Add a legend to the figure, not the subplot, only once after the loop
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right')

# Display the plot
plt.show()
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages','English','Communications and Journalism','Art and Performance','Social Sciences and History']
other_cats = ['Health Professions','Public Administration','Education','Agriculture','Business','Architecture']

years = np.arange(1970, 2011)
fig = plt.figure(figsize=(18, 10))

def plot_chart(ax, category, degrees, topmost, bottommost):

    ax.plot(years, 100 - degrees[category][:len(years)], c='orange', label='Men',linewidth=3)
    ax.plot(years, degrees[category][:len(years)], c='blue', label='Women',linewidth=3)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)

    # ax.set_xlim(1970, 2010)
    # ax.set_ylim(0, 100)
    ax.set_title(category)

    ax.tick_params(labelbottom=True)
    for spine in ax.spines.values():
        spine.set_visible(False)
    # iu chnh v tr ch thch da theo topmost v bottommost
    if topmost:
    # Chnh sa v tr ch thch ca plot 1 hng 1 ct 1
        ax.text(2005, degrees[category].iloc[-1] + 5, 'Women')  # i 'Women' thnh 'Men'
        ax.text(2005, 100 - degrees[category].iloc[-1] - 10, 'Men')  # i 'Men' thnh 'Women'

    if bottommost and (col == 0 or col == 2):  # Thm ch thch cho ct th nht v th ba
        ax.text(2005, degrees[category].iloc[-1] +20, 'Men')  # i 'Men' thnh 'Women'
        ax.text(2005, 100 - degrees[category].iloc[-1] - 20, 'Women')  # i 'Woman' thnh 'Men'

num_rows, num_cols = 6, 3

for col in range(num_cols):
    if col == 0:
        cat_list = stem_cats
    elif col == 1:
        cat_list = lib_arts_cats
    else:
        cat_list = other_cats

    for row in range(num_rows):
        ax = fig.add_subplot(num_rows, num_cols, (row * num_cols) + (col + 1))

        if row < len(cat_list):
            cat = cat_list[row]
            topmost = (row == 0)
            bottommost = (row == len(cat_list) - 1) or (col == 1 and row == len(cat_list) - 1)
            plot_chart(ax, cat, women_degrees, topmost=topmost, bottommost=bottommost)
            ax.set_title(cat)

            # Ch hin th nhn 0 v 100 trn trc y
            ax.set_yticks([0, 100])

            # iu chnh x ticks v x labels
            if bottommost:
                ax.tick_params(labelbottom=True)
            else:
                ax.tick_params(labelbottom=False)  # Xa nhn trc x nu khng phi l plot cui cng
                ax.tick_params(bottom=False)  # Xa cc tick marks trn trc x

        else:
            ax.axis('off')
plt.tight_layout()
plt.savefig('gender_degrees.png')  # Lu biu  trc khi hin th
```
